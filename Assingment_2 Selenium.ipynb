**Question 1**- Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.e webpage https://www.shine.com/

```python
# import necessary library
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
# open the shine page on automated chrome browser
driver.get("https://www.shine.com/")
```

```python
# entering desigation and locations as required in the question
designation = driver.find_element(By.XPATH,'//input[@id="id_q"]')
designation.send_keys("Data Analyst")
```

```python
location = driver.find_element(By.XPATH,'//input[@id="id_loc"]')
location.send_keys("Bangalore") # location
```

```python
search = driver.find_element(By.XPATH,'//*[@id="__next"]/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button') # scrap xpath of search button
search.click()
```

```python
#empty enumerate for store data
job_titles = []
job_locations = []
company_names = []
experience_required = []
```

```python
#scraping title
titles_tags = driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
for i in titles_tags:
    job_titles.append(i.text)
    job_titles[:10]
# scraping location
location_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for a in location_tags:
    job_locations.append(a.text)
    job_locations[:10]
#scraping company name
company_tags = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for b in company_tags:
    company_names.append(b.text)
    company_names[:10]
# scraping experience years    
experience_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for c in experience_tags:
    experience_required.append(c.text)
    experience_required[:10]
    
    
    
    
```

```python
print(len(job_titles[:10]),len(job_locations[:10]),len(company_names[:10]),len(experience_required[:10]))
```

```python
#Create DataFrame
data=pd.DataFrame({})
data['Job_title']=job_titles[:10]
data['Locations']=job_locations[:10]
data['Company']=company_names[:10]
data['Experience']=experience_required[:10]
data
```
